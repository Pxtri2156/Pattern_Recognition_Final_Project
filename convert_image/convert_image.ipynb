```python
import numpy as np
import pandas as pd
import cv2
```

```python
a = np.load('/content/drive/MyDrive/Emotion_Speech_Recognition/RAW_DATASET/IEMOCAP/Train_IEMOCAP.npy')
b = np.load('/content/drive/MyDrive/Emotion_Speech_Recognition/RAW_DATASET/IEMOCAP/Dev_IEMOCAP.npy')
```

```python
print(a)
```

```python
image_id = []
labels = []
fold = []
```

```python
cp -r /content/drive/MyDrive/Emotion_Speech_Recognition/image.zip image.zip
```

```python
ls /content/IEMOCAP_IMAGE |wc -l
```

```python
rm -rf train
```

```python
mkdir train
```

```python
root = '/content/IEMOCAP_IMAGE/'
save = '/content/train/'
```

```python
stt = 0
for i in a:
  image = i.split('.')[0].split('/')[-1]
  label = image.split('_')[-1]
  image_id.append(image + '.png')

  path = root + image + '.png'
  out = save + image + '.png'

  img = cv2.imread(path)
  cv2.imwrite(out,img)
  stt += 1
  if stt % 100 == 0:
    print(stt, '/',len(a))

  fold.append(1)
  labels.append(label)
```

```python
stt = 0
for i in b:
  image = i.split('.')[0].split('/')[-1]
  label = image.split('_')[-1]
  image_id.append(image + '.png')

  path = root + image + '.png'
  out = save + image + '.png'

  img = cv2.imread(path)
  cv2.imwrite(out,img)
  stt += 1
  if stt % 100 == 0:
    print(stt, '/',len(b))

  fold.append(0)
  labels.append(label)
```

```python
df = pd.DataFrame(data=image_id,columns=['image_id'])
df['class'] = labels
df['fold'] = fold
```

```python
df.to_csv('train.csv', index=False)
```

```python
import pandas as pd
content = pd.read_csv('/content/submission_res.csv')
labels = []
for img in content["image_id"]:
  label = img.split('.')[0].split('_')[-1]
  labels.append(label)

categ = {
    "ang": 0,
    "exc": 1,
    "fru": 2,
    "hap": 3,
    "neu": 4,
    "sad": 5
}

content['ground_truth'] = labels
content.to_csv('final_res.csv',index=False)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
con = pd.read_csv('/content/final_res.csv')
con['ground_truth'] = con.apply(lambda row: categ[row["ground_truth"]], axis =1)

val_Y = []
preds = []

for lb in con['ground_truth']:
  val_Y.append(lb)

for lb in con['label']:
  preds.append(lb)
```

```python
print(classification_report(val_Y, preds))
```
